```python
# Step 8 read the file 'gapminder_population.xlsx' into pandas.
pop=pd.read_excel('\\Users\\X1\\Documents\\spiced_projects\\fenugreek-student-code\\week1\\data\\gapminder_population.xlsx', index_col=0)
pop.head()
```

```python
# the name of the index should be called country. change the name of the index
pop.index.name='country'
```

```python
# convert the table into long format. First, we have to move the row index into a column
pop.reset_index(inplace=True)
```

```python
# we use the convenient melt function for the conversion
pop=pop.melt(id_vars='country', var_name='year', value_name='population')
pop.head()
```

```python
#  read the file 'continents.csv' into pandas.
cont=pd.read_csv('\\Users\\X1\\Documents\\spiced_projects\\fenugreek-student-code\\week1\\data\\continents.csv',sep=';', index_col=0)
cont.head()
```

```python
# check the shape of both tables
cont.shape
```

```python
# convert the table into long format. First, we have to move the row index into a column
cont.reset_index(inplace=True)
```

```python
# we use the convenient melt function for the conversion
cont=cont.melt(id_vars='country', value_name='continent')
cont.head()
```

```python
# we can start gluing the DataFrames together
#first of all, we merge fert and pop to get df_1
df_1=fert.merge(pop, how='inner', on=['country','year'])
df_1
```

```python
#now we merge population to df_1 table
df_2=df_1.merge(life, how='inner', on=['country', 'year'])
df_2
```

```python
#then we merge continent to df_2 table
df=df_2.merge(cont, how='inner', on='country')
df.drop('variable', axis=1, inplace=True)
df
```

```python
df.dropna(inplace=True)
```

```python
# Step 10 With all the data merged into a single object, you can now select specific columns and plot them
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="dark")
plt.style.use('dark_background')
plt.figure(figsize=(8,5))
df_subset = df.loc[df['year'] == 2000]
sns.scatterplot(x='life_expectancy', y='fertility_rate',data=df_subset, alpha=0.5,hue='continent',legend=False)


plt.axis((0,100,0,15))
plt.title('Correlation between fertility and life expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Fertility')
    
sns.despine()

```

```python
# Step 12 Add another dimension to the visualization by utilizing the population data.
#Adjust the plot, so that the size of the symbols represent the population from the gapminder dataset.
 
for i in range(1960,2016):
    sns.set_theme(style="dark")
    plt.style.use('dark_background')
    plt.figure(figsize=(8,5))
    df_subset = df.loc[df['year'] == i]
    sns.scatterplot(x='life_expectancy', y='fertility_rate',size=df_subset['population']//100_000, legend=False, hue='continent',
            data=df_subset, alpha=0.5)
    
    fname='lifeexp_'+str(i)+'.png'
    
    plt.axis((0,100,0,15))
    plt.title(f'Correlation between fertility and life expectancy-{i}')
    plt.xlabel('Life Expectancy')
    plt.ylabel('Fertility')
    
    sns.despine()
    plt.savefig(fname, dpi=90)
    plt.close()
```

```python
# Step 11 To create an animated scatterplot showing the correlation between life expectancy and fertility, create one plot for each year between 1960 and 2015
for i in range(1960,2016):
   
    sns.set_theme(style="dark")
    plt.style.use('dark_background')
    plt.figure(figsize=(8,5))
    df_subset = df.loc[df['year'] == i]
    sns.scatterplot(x='life_expectancy', y='fertility_rate',hue='continent', legend=False,
            data=df_subset, alpha=0.5)
    
    fname='lifeexp_'+str(i)+'.png'
    
    plt.axis((0,100,0,15))
    plt.title(f'Correlation between fertility and life expectancy-{i}')
    plt.xlabel('Life Expectancy')
    plt.ylabel('Fertility')
    
    sns.despine()
    plt.savefig(fname, dpi=90)
    plt.close()
```
