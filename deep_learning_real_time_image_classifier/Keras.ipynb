# Build a model in Keras

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X,y = make_moons(n_samples=5000, noise=0.2, random_state=42)
```

```python
X.shape , y.shape
```

```python
plt.scatter(X[:,0], X[:,1], c=y, alpha=0.5)
```

## Import Keras components

```python
from tensorflow.keras.models import Sequential # Feed Forward Network in Keras Terminology
from tensorflow.keras.layers import Dense # Fully connected layers
from tensorflow.keras.layers import Activation 
from tensorflow.keras import backend as K
```

```python
#clear Keras backend session
K.clear_session()
```

## Define the model

```python
#input shape
X.shape[1]
```

```python
model_1 = Sequential([
    # hidden layer, as the very first layer
    Dense(2, input_shape = (2,)), # the shape of ONE observation i.e. the number of features
    Activation('sigmoid'),
    # output layer
    Dense(1), 
    Activation('sigmoid')
    
])
```

```python
model_1.summary()
```

```python
model_1.get_weights()
```

## Compile the model

```python
model_1.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

## Fit the model

```python
history = model_1.fit(X,y, epochs = 500, batch_size = 50, validation_split = 0.2, verbose = 1)
```

```python
model_1.get_weights()
```

## Evaluate the model

```python
plt.plot(history.history['accuracy'], label = 'train_data')
plt.plot(history.history['val_accuracy'], label = 'validation_data')
plt.legend()
```

```python
plt.plot(history.history['loss'], label = 'train_data')
plt.plot(history.history['val_loss'], label = 'validation_data')
plt.legend()
```

## Take a look at the prediction

```python
ypred = model_1.predict(X)
```

```python
plt.hist(ypred)
```

```python
plt.scatter(X[:,0],X[:,1], c = ypred, alpha = 0.5)
```
